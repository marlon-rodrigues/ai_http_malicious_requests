```python
import pandas as pd
normal_requests = pd.read_csv('Datasets/normal_requests_csv.csv', header=None)
normal_requests.head(5)
```

```python
normal_requests['label_id'] = 0
normal_requests['label_desc'] = "normal"
normal_requests.head(5)
```

```python
len(normal_requests)
```

```python
malicious_requests = pd.read_csv('Datasets/malicious_payload_csv.csv', header=None, sep='\t')
malicious_requests.head(5)
```

```python
len(malicious_requests)
```

```python
malicious_requests['label_id'] = 1
malicious_requests['label_desc'] = "anomaly"
malicious_requests.head(5)
```

```python
combined_dataset = pd.concat([normal_requests, malicious_requests])
combined_dataset.columns = ['payload', 'label_id', 'label_desc']
combined_dataset
```

```python
# Get number of clean and malicious requests
grouped_by_label = combined_dataset.groupby(['label_desc']).count()
grouped_by_label
```

```python
# Get percentage of malicious requests in dataset
print("Malicious request logs in dataset: {:0.2f}%".format(float(grouped_by_label['payload'][0]) / len(combined_dataset) * 100))
```

```python
# Save dataframe to csv
combined_dataset.to_csv('Datasets/all_requests_csv.csv', index=False, encoding='utf-8')
```
